# Conditions and cycles


### if condition

```python
# minimal example

a = 1
if a > 0:
    print("a is positive")
```

```python
# full example if-elif-else

if (a == 1):
    print("hit!")
    a = 1000
    print(a)
    if (a == 1000):
        print('double hit!')
    print('ahoj')
elif (a == 0):
    print("bad luck")
elif (a == 2345):
    a = 0
    print(a)
else:
    print("try another time")

```

### while cycle

```python
i = 1
while i < 10:
    print(i)
    i += 1

```

### for cycle

```python
for i in range(10):
    if i % 2 == 0:
        print(i)
```

### Important words

- *break*: breaks the nearest enclosing loop
- *continue*: goes to the next iteration of the nearest enclosing loop
- *pass*: does nothing, placeholder for future code
