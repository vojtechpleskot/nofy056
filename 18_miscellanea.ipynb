# git

- Tool for version control and collaboration on code projects.

- Useful for tracking changes in code, and managing different versions of a project.

- Common web platforms: GitHub, GitLab, and others.

- The code is easy to share using these platforms.

- Source code can be managed using git, and LaTeX documents also.

# Basic terminology

- Repository (repo): A storage space for your project, which can be local or hosted on a platform like GitHub.

- Clone: A copy of a repository that you can work on locally.

- Branch: A separate line of development in a repository, allowing you to work on different features or fixes without affecting the main codebase.

- Commit: A snapshot of changes made to the files in the given branch and repository.

- Push: Uploading your local commits to a remote repository.

- Pull or merge: Downloading changes from a remote repository to your local copy.

- Checkout: Switching between different branches in your repository.

# Working with git

- Install git on your local machine.

- You can manage your git projects in Visual Studio Code using the built-in git support.

- You can also use git from the command line.

- Basic workflow (if collaborating with others):

  - Clone a repository from GitHub, GitLab, or another platform to your local machine.

  - Create a new branch for your work.

  - Make changes to the files in your local copy.

  - Commit your changes with a descriptive message.

  - Push your commits to the remote repository on GitHub.

  - Create a pull request to merge your changes into the main branch.

- Basic workflow (if working alone):

  - Create a new repository on GitHub, GitLab, or another platform.

  - Clone the repository to your local machine.

  - Add files and make changes to the files in your local copy.

  - Commit your changes with a descriptive message.

  - Push your commits to the remote repository on GitHub.
