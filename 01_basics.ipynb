# Useful operators

- a = 9 assignment of value 9 to variable a
- a == 9, a > 9, a < 9, a >= 9, a <=9 comparisons of the value stored in a with 9
- a += 1 incrementing the value of a by one
- a -= 3 lowering the value of a by 3
- a *= 2 doubling the value of a
- a /= 7 dividing the value of a by 7
- a %= 5 getting the remainder of a divided by 5
- a //= 11 performing floor division of a by 11
